```python
def precision_recall_at_k(predictions, k=10, threshold=3.5):
    user_est_true = defaultdict(list)
    for pred in predictions:
        user_est_true[pred.uid].append((pred.est, pred.r_ui))
    #print(user_est_true)
    precisions = []
    recalls = []

    for user_id, user_ratings in user_est_true.items():
        # Sort user's predictions by estimated rating
        user_ratings.sort(key=lambda x: x[0], reverse=True)
        top_k_ratings = user_ratings[:k]

        # Count how many of the top-K are actually relevant (true rating >= threshold)
        # n_rel = sum((true_r >= threshold) for (_, true_r) in user_ratings)
        # n_rec_k = sum((est >= threshold) for (est, _) in top_k_ratings)
        # n_rel_and_rec_k = sum(((true_r >= threshold) and (est >= threshold)) for (est, true_r) in top_k_ratings)
        n_rel = 0  # All the rows where true ratings where greater than 3.5
        for _, true_rating in user_ratings:
            if true_rating >= threshold:
                n_rel += 1

        n_rec_k = 0  #All the predicted ratings where greater than 3.5
        for est_rating, _ in top_k_ratings:
            if est_rating >= threshold:
                n_rec_k += 1

        n_rel_and_rec_k = 0
        for est_rating, true_rating in top_k_ratings:
            if est_rating >= threshold and true_rating >= threshold:
                n_rel_and_rec_k += 1



        if n_rec_k != 0:
            prec = n_rel_and_rec_k / n_rec_k
        else:
            prec = 0

        if n_rel != 0:
            rec = n_rel_and_rec_k / n_rel
        else:
            rec = 0


        precisions.append(prec)
        recalls.append(rec)

    avg_precision = sum(precisions) / len(precisions)
    avg_recall = sum(recalls) / len(recalls)

    return avg_precision, avg_recall

# Compute Precision@K and Recall@K
precision, recall = precision_recall_at_k(predictions, k=10, threshold=3.5)

print(precision)
print(recall)
```
